```python
import joblib

# Load the saved model and vectorizer
svm_model = joblib.load('/content/drive/MyDrive/AIML/sentemental_analysis/svm_model.pkl')
tfidf = joblib.load('/content/drive/MyDrive/AIML/sentemental_analysis/tfidf_vectorizer.pkl')

def get_sentiment(text, model, vectorizer):
    clean_input = text.lower()
    input_vector = vectorizer.transform([clean_input]).toarray()
    prediction = model.predict(input_vector)[0]
    return prediction.capitalize()

# Run the app
if __name__ == "__main__":
    user_input = input("Enter your message: ")
    predicted_sentiment = get_sentiment(user_input, svm_model, tfidf)
    print("\nPredicted Sentiment:", predicted_sentiment)
```
